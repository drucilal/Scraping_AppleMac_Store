## Size

#### Group the dataset by size to visaulize the smallest to largest apps
GB: gigabytes (largest),
MB: megabytes ,
KB: kilobytes (smallest),

#### Grouping by SIze

Large Apps : GB

```python
large = mac_apps[mac_apps['size'].str.contains('GB')]
large_size = large[['category','name','size', 'price', 'total_languages']].sort_values(by='size', ascending = False)
large_size['size'] = large_size['size'].str.replace('GB', '')
large_size['size'] = large_size['size'].astype(float)
```

Apps : MB

```python
medium = mac_apps[mac_apps['size'].str.contains('MB')]
medium_size = medium[['category','name','size', 'price', 'total_languages']].sort_values(by='size', ascending = False)
medium_size['size'] = medium_size['size'].str.replace('MB', '')
medium_size['size'] = medium_size['size'].astype(float)
```

Smallest Apps : KB

```python
small = mac_apps[mac_apps['size'].str.contains('KB')]
small_size = small[['category','name','size', 'price', 'total_languages']].sort_values(by='size', ascending = False)
small_size['size'] = small_size['size'].str.replace('KB', '')
small_size['size'] = small_size['size'].astype(float)
```

### Given that graphic apps contained the most memory, could it be the most expensive category as well? 

### First,  let's look and visualize free and paid apps. 

```python
price_c = mac_apps['price_category'].value_counts().reset_index()
price_c.rename(columns={'index':'price_category', 'price_category':'count'}, inplace=True)
```

```python
sns.set(style = 'whitegrid', font_scale = 1.5, rc={"figure.figsize": (5, 10), 'grid.color': '0.9'},)
s = sns.barplot(x="price_category", y="count", data=price_c, label=" ", palette = "Blues_d", ci = None)
plt.xlabel('Free or Paid App ') 
plt.ylabel('Count')
plt.show(s)
```

Findings : Majority of apps in the mac store are paid. To continue to answer our question, let's look paid apps in each category

```python
# Filter paid vs free
paid = mac_apps.query('price_category == 1')
free = mac_apps.query('price_category == 0')
```

```python
sns.set(style = 'whitegrid', font_scale = 1.8, rc={"figure.figsize": (12, 13), 'grid.color': '0.9'},)
p = sns.countplot(y = 'category', data = paid, palette= "Blues_d", order = paid['category'].value_counts().index)
plt.xlabel('Number of Paid Apps') 
plt.ylabel('Category')
plt.show(p)
```

Findings: Majority of paid apps are from the graphics and design category. Given that graphics and design apps had the highest number of large sized apps, it would make sense that majority of the apps were paid. 

```python
# Total Price 
paid.groupby('category', as_index=True)['price'].sum().sort_values().plot(kind='barh', color = '#34495e')
```

Findings: Again we see that the most expensive apps are Graphics & Design apps. 
Lets' dig into this genre and visualize the most expensive apps. 

```python
graph = paid[paid['category'].str.contains('Graphics & Design')].sort_values(by ='price', ascending = False)
paid_graph =  graph.head(10)
sns.set(style = 'whitegrid', font_scale = 1.5, rc={"figure.figsize": (7, 12), 'grid.color': '0.9'},)
sns.barplot(x="price", y="name", data= paid_graph, label=" ", color="blue", ci = None)
```

Findings: Shark Cad is the most expensive graphics and design apps ($700). This app consists of various graphic programs and features that enable users to work more efficiently. 

For the affect on price vs memory size, we saw how graphic apps consisted of the most memory and was the most expensive.
